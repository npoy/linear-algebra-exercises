##### Exercise 1 (Video: Code Challenge: Geometric transformations via matrix multiplications):

- Generate XY coordinates for a circle
- Plot the circle
- Create a 2x2 matrix (starting with I (identity matrix))
- Multiply matrix by coordinates
- Plot the new coords
- Try with various matrices
- Try with a singular matrix (columns form a linearly dependent set)

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, +np.pi, 100)
xy = np.vstack((np.cos(x), np.sin(x))).T
# xy = np.array((np.cos(x), np.sin(x))).T

# print(x)
# print(xy)
print(np.shape(xy))

# plot the circle
plt.plot(xy[:,0], xy[:,1], 'o')

# Create a 2x2 matrix
T = np.array([[1,2], [2,1]])

# multiply matrix by coords
newxy = xy@T

# plot the new coords
plt.plot(newxy[:,0], newxy[:,1], 'o')

plt.axis('square')
plt.show()
```
