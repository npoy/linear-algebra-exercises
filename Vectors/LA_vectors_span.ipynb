##### Exercise 8 (Video: Is in the span?):
### In the span?

- The vector below is contained in the span of one of the following sets. But which one is it?!?
[1 2 3]

1. {[1 2 4],[3 2 5]}
2. {[1 0 0],[0 1 0]}
3. {[2 6 -3],[0 -2 9]}

Answer: 3

```python
import numpy as np
import matplotlib.pyplot as plt

# set S1
# S1 = np.array([1, 2, 4])
# S2 = np.array([3, 2, 5])

# set S2
# S1 = np.array([1, 0, 0])
# S2 = np.array([0, 1, 0])

# set S3
S1 = np.array([2, 6, -3])
S2 = np.array([0, -2, 9])

# vectors v and w
v = np.array([1, 2, 3])

# draw vectors
fig = plt.figure()
# ax = fig.add_subplot(projection='3d')
ax = plt.axes(projection = '3d')
ax.plot([0, S1[0]],[0, S1[1]],[.1, S1[2]+.1],'r',linewidth=1)
ax.plot([0, S2[0]],[0, S2[1]],[.1, S2[2]+.1],'r',linewidth=1)

ax.plot([0, v[0]],[0, v[1]],[.1, v[2]+.1],'g',linewidth=1)

# now draw plane
xx, yy = np.meshgrid(range(-15,16), range(-15,16))
cp = np.cross(S1,S2)
z1 = (-cp[0]*xx - cp[1]*yy)*1./cp[2]
ax.plot_surface(xx,yy,z1)

plt.show()
```
